```python
data["id"] = data["id"].astype("float") #id nin türünü değiştiriyoruz.
```

```python
ortalama=sum(data.vote_average)/len(data.vote_average) #oyun ortalamasını buluyoruz.
data["ortalama_oyun_seviyesi"]=[ "yüksek" if i >ortalama else "alçak" for i in data.vote_average] #ortalama oyun altında kalan alçak üstünde olan yüksek
data.loc[:20,["original_title","ortalama_oyun_seviyesi"]] #ilk 20 filmin ortalama oyun altında kalıp kalmadığına bakıyoruz. 
```

<h2>TOP 6 MOVİE </h2>

Let's visualize the data we analyzed

```python
data.plot(grid=True, alpha=0.7) #data hakkında genel bir grafik çizdirelim.
```

```python
data1=data.loc[2000:,["popularity","budget"]]
data1.plot(subplots=True) #subplots ayrı grafikler çizdirmeye yarıyor.
plt.show()
```

```python
data.boxplot(column="vote_average",by="ortalama_oyun_seviyesi") #oy ortalamalarında ne kadar aşırı değer var görelim.
```

```python
data.runtime.plot(kind="line",color="red",label="runtime",grid=True,alpha=0.8,linewidth=1,linestyle=":") #filmin populeritesini ve oynatma süresini karşılaştırdım 
data.popularity.plot(color="green",label="popularity",grid=True,alpha=0.6,linestyle="-.") #üst satırda ve burda grafigin türünü rengini ve bir özelliğini belirledim.
plt.legend() #labelların en uygun yere yerleşmesini sağlar.
plt.xlabel("runtime") # x sutunun neyi temsil ettiğini gösterir.
plt.ylabel("popularity") #düzlemin y sutunun neyi gösterdiğini gösteren ve hemen y sutunun yanında yazılı olan.
plt.title("Runtime-Popularity") #tablonun en üstüne yerleşecek başlık
plt.show()
```

```python
data.plot(kind="scatter",x="budget",y="revenue",alpha=0.7,color="purple") #plotun türü,x sutunun verisinin ne olacagı,y sutunun verisinin ne olacagı,şeffaflıgı,rengi
plt.legend()
plt.xlabel("budget")
plt.ylabel("revenue")
plt.title("Budget-Revenue")
plt.show()
```
